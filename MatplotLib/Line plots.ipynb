```python
# let's style our plot using a style instead 
```

```python
plt.style.available
```

```python
plt.style.use("grayscale")
max_attack.plot(kind="barh")
```

```python
plt.style.use("dark_background")
max_attack.plot(kind="barh")
```

```python
def all_designs():
    for style in plt.style.available:
        plt.style.use(style)
        max_attack.plot(kind="barh", figsize=(4,3), fontsize=6)
        print(style)
        plt.show()

all_designs()
        
```

```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
air_quality = pd.read_csv("gap.csv", usecols=["Country","City","AQI Value","AQI Category"])
air_quality.sort_values(by="AQI Value", inplace= True)
air_quality
```

```python
# First, let's group each country 
```

```python
air_quality = pd.read_csv("gap.csv",usecols=["Country","City","AQI Value", "AQI Category"])
air_quality.sort_values(by="AQI Value", inplace=True)
air_quality 
```

```python
countries = air_quality.groupby("Country")
countries.groups 
```

```python
average = countries.mean(numeric_only=True).round(2)
average.sort_values(by="AQI Value", inplace=True)
average
```

```python
def color(number: int) -> str:
    if number < 50:
        return "forestgreen"
    elif number < 100:
        return "orange"
    else:
        return "crimson"
        
```

```python
# Applying this function to each row so we get a colour for it 
```

```python
colors = average["AQI Value"].apply(color)
colors 
```

```python
average.plot(y="AQI Value",
             kind="barh",
             color=colors.values,
             figsize=(10,40),
             title="Air pollution in country")
```

```python
# We can also do it on a specific country
```

```python
Palau = countries.get_group("Palau")
Palau
```

```python
Peru = countries.get_group("Peru")
Peru
```

```python
colors = Peru["AQI Value"].apply(color)
colors 
```
